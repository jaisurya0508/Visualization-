```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y_views=[23,45,67,12,9,9,7,56,45,34,89,90,67,56]

x=range(1,15)
miny=min(y_views)
x_co=x[y_views.index(miny)]
plt.plot(x,y_views,label="u tube views")

plt.xlabel('days')
plt.ylabel('Views')
plt.grid(True)
plt.title("You tube viewers result")
plt.legend(loc='upper right')
plt.annotate('Low', xy=(x_co,miny),xytext=(5,100),arrowprops=dict(facecolor='red',shrink=0.5),)
plt.show()
```

### using style sheets

```python
y_views=[23,45,67,12,9]
f_views=[13,25,67,62,99]
x=range(5)
print(plt.matplotlib.style.available)
#plt.use('seaborn')
```

```python
plt.matplotlib.rcParams['font.family']='fantasy'
plt.bar([a-0.25 for a in x],y_views,label="u tube views",color='b',width=0.25,linestyle='dashed',linewidth=3,align='edge')
plt.bar([a+0.25 for a in x],f_views,label="facebook views",color='r',width=0.25,linestyle='dashed',linewidth=3,align='edge')
plt.xlabel('Days')
plt.ylabel('Views')
plt.title("Social media viewers result")
plt.legend(loc='upper right')
plt.grid(True,linewidth=1,color='b',linestyle='dashed')
plt.xticks(x)
```
