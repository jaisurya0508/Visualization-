### plotting data from list

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
views=[700,789,456,240,320,569,100]
days=range(1,8)
#plt.plt(x,y)
plt.plot(days,views)
plt.show()
```

### labels and legend

```python
plt.xlabel('Days')
plt.ylabel('Views')
plt.plot(days,views,label="u tube views")
plt.title("You tube viewers result")
plt.legend(loc='lower left')
plt.show()
```

### colors and fonts

```python
plt.xlabel('Days')
plt.ylabel('Views')
plt.plot(days,views,label="u tube views",color='r',marker='o',markerfacecolor='b')
plt.title("You tube viewers result")
plt.legend(loc='lower left')
plt.show()
```

### Line style and width

```python
plt.xlabel('Days')
plt.ylabel('Views')
plt.plot(days,views,label="u tube views",color='r',marker='o',markerfacecolor='b',linestyle='dashed',linewidth=3)
plt.title("You tube viewers result")
plt.legend(loc='upper right')
plt.show()
```

### Multiple plot

```python
y_views=[23,45,67,12,9]
f_views=[90,56,45,22,10]
days=range(5)
plt.xlabel('Days')
plt.ylabel('Views')
plt.plot(days,y_views,label="u tube views",color='r',marker='o',markerfacecolor='b',linestyle='dashed',linewidth=3)
plt.plot(days,f_views,label="FB views",color='b',marker='o',markerfacecolor='y',linestyle='dashed',linewidth=3)
plt.title("Social Media viewers result")
plt.legend(loc='upper right')
plt.show()
```

### Setting x,y axis limits

```python
y_views=[23,45,67,12,9]
f_views=[90,56,45,22,10]
days=range(5)
plt.xlabel('Days')
plt.ylabel('Views')
plt.plot(days,y_views,label="u tube views",color='r',marker='o',markerfacecolor='b',linestyle='dashed',linewidth=3)
plt.plot(days,f_views,label="FB views",color='b',marker='o',markerfacecolor='y',linestyle='dashed',linewidth=3)
plt.title("Social Media viewers result")
plt.legend(loc='upper right')
plt.xlim(1,5)
plt.ylim(0,100)
plt.show()
```

### Adding gridlines to plot

```python
y_views=[23,45,67,12,9]
f_views=[90,56,45,22,10]
days=range(5)
plt.xlabel('Days')
plt.ylabel('Views')
plt.plot(days,y_views,label="u tube views",color='r',marker='o',markerfacecolor='b',linestyle='dashed',linewidth=3)
plt.plot(days,f_views,label="FB views",color='b',marker='o',markerfacecolor='y',linestyle='dashed',linewidth=3)
plt.title("Social Media viewers result")
plt.legend(loc='upper right')
plt.grid(True,linewidth=3,color='Black',linestyle='-')
plt.xlim(1,5)
plt.ylim(0,100)
plt.show()
```
