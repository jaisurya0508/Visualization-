```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y_views=[23,45,67,12,9]
f_views=[13,25,67,62,99]
x=range(5)
plt.xlabel('Days')
plt.ylabel('Views')
plt.bar([a-0.25 for a in x],y_views,label="u tube views",color='b',width=0.25,linestyle='dashed',linewidth=3,align='edge')
plt.bar([a+0.25 for a in x],f_views,label="fb views",color='r',width=0.25,linestyle='dashed',linewidth=3,align='edge')
plt.title("You tube viewers result")
plt.legend(loc='upper right')
plt.grid(True,linewidth=1,color='b',linestyle='dashed')
plt.xticks(x)


plt.show()
```

### Histogtam

```python
y=np.arange(20)

bins=[1,2,3,4,5]
plt.hist(y,bins,label='Result of viewers',color='b')
plt.xlabel('day')
plt.ylabel('Views')
plt.legend()
plt.grid(True,linewidth=1,linestyle='dashed')
plt.title('daily viewers')
plt.xticks(bins)
plt.show()
```

### piechart

```python
labels=['facebook','youtube','Instagram','Whatsapp']
views=[234,567,890,784]

plt.pie(views,labels=labels)
plt.show()
```

```python
labels=['facebook','youtube','Instagram','Whatsapp']
views=[234,567,890,784]
explode=[0,0,0.2,0]
plt.pie(views,labels=labels,explode=explode)
plt.show()
```

```python
labels=['facebook','youtube','Instagram','Whatsapp']
views=[234,567,890,784]
explode=[0,0,0.2,0]
plt.pie(views,labels=labels,explode=explode,autopct='%1.2f%%')
plt.show()
```

```python
labels=['facebook','youtube','Instagram','Whatsapp']
views=[234,567,890,784]
explode=[0,0,0.2,0]
plt.pie(views,labels=labels,explode=explode,autopct='%1.1f%%',wedgeprops={'width':0.3})
plt.show()
```

```python
plt.pie(views,labels=labels,explode=explode,autopct='%1.1f%%',shadow=True,wedgeprops={'width':0.3})
plt.show()
```
