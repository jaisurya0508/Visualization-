### Scatter

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y_views=[23,45,67,12,9]
days=range(5)
plt.xlabel('Days')
plt.ylabel('Views')
plt.scatter(days,y_views,label="u tube views",color='r',linestyle='dashed',linewidth=3)
plt.title("You tube viewers result")
plt.legend(loc='upper right')
plt.grid(True,linewidth=1,color='b',linestyle='-')
plt.xlim(1,5)
plt.ylim(0,100)
plt.savefig('Canada.jpg',dpi=500,facecolor='b')
plt.show()
```
